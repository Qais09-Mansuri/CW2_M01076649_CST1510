# **QUESTION 9:**
Write a guessing game in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.

```python
#ANSWER 9
import random  # importing the random module to generate random numbers

# generating a secret number between 0 and 100
secret_number = random.randint(0, 100)

# variable to count how many tries the user takes
tries = 0

# variable to store the user's guess
guess = None

# loop keeps running until the user guesses the secret number
while guess != secret_number:
    guess = int(input("Guess the number: "))  # asking user to guess
    tries += 1  # incrementing the number of tries

    # giving feedback if the guess is too low
    if guess < secret_number:
        print("Number is too small")
    
    # giving feedback if the guess is too high
    elif guess > secret_number:
        print("Number is too big")
    
    # if the guess is correct
    else:
        print(f"Congrats! You guessed it in {tries} tries")
```

# **QUESTION 10:**

Use 0 as a sentinel value to mark the end of the input. Once all of the values have been read your program should display them (except for the 0) in reverse order, with one value appearing on each line.

```python
#ANSWER 10
# creating an empty list to store numbers
numbers = []

# repeatedly ask the user to enter numbers until they enter 0
while True:
    num = int(input("Enter an integer (0 to stop): "))  # take input from user
    if num == 0:  # if user enters 0, stop the loop
        break
    numbers.append(num)  # add the entered number to the list

# print a message before showing numbers in reverse
print("\nNumbers in reverse order:")

# loop through the list in reverse and print each number
for n in reversed(numbers):
    print(n)
```

For example:
Some of the palindrome words are civic, radar, level, rotor, kayak, madam, refer, rotator, deified, racecar and reviver.
The function should one string, and return True it is palindrome, otherwise False.

```python
#ANSWER 11
# defining a function to check if a string is a palindrome
def p(s):
    return s == s[::-1]  # compare the string with its reverse

# asking the user to enter a word
text = input("Enter a word: ")

# printing True if the word is a palindrome, False otherwise
print(p(text))
```

# **QUESTION 12:**
Your function will take three parameters: the list, the minimum value and the maximum value. It will return an integer result.
Include a main program that demonstrates your function for several different lists, minimum values and maximum values. Ensure that your program works correctly for both lists of integers and lists of floating-point numbers. (For clarification.
Count the number of elements in a list that are greater than or equal to some minimum value and less than some maximum value.)

```python
#ANSWER 12
# defining a function to count numbers within a specific range
def countRange(numbers, min_val, max_val):
    count = 0  # initialize counter
    for n in numbers:  # loop through each number in the list
        if min_val <= n < max_val:  # check if the number is within the range
            count = count + 1  # increment the counter
    return count  # return the total count

# example list of integers
nums1 = [1, 5, 8, 10, 15, 20]

# example list of floating point numbers
nums2 = [2.5, 3.6, 7.2, 9.8, 12.0]

# counting numbers in nums1 between 5 (inclusive) and 15 (exclusive)
print(countRange(nums1, 5, 15))  

# counting numbers in nums2 between 3.0 (inclusive) and 17.0 (exclusive)
print(countRange(nums2, 3.0, 17.0))
```

Create a list of passwords and apply the function on the list of passwords

```python
#ANSWER 17
# Function to count different types of characters in a string
def count_characters(s):
    # Initialize counters for lowercase, uppercase, digits, and special characters
    lower = upper = digit = special = 0

    # Loop through each character in the string
    for char in s:
        if char.islower():      # check if character is lowercase
            lower += 1
        elif char.isupper():    # check if character is uppercase
            upper += 1
        elif char.isdigit():    # check if character is a digit
            digit += 1
        else:                   # any other character is special
            special += 1

    # Print the results for the current string
    print(f"\nPassword: {s}")
    print("Lowercase letters:", lower)
    print("Uppercase letters:", upper)
    print("Digits:", digit)
    print("Special symbols:", special)

# List of sample passwords to analyze
passwords = ["P@#yn26at^&i5ve", "Hello123!", "MDX@2025", "qaismansuri09"]

# Apply the function to each password in the list
for pw in passwords:
    count_characters(pw)
```

# **QUESTION 18:**
Password Validation with Multiple Criteria: Ask the user for a password and validate whether it:
Has more than 8 characters.
Contains at least one uppercase letter.
Contains at least one lowercase letter.
Contains at least one number.
Contains at least one special character (@, #, $, etc.).
Extra: If the password contains a space, print an error message

```python
#ANSWER 18
# Function to validate a password based on multiple conditions
def validate_password(password):
    # Check for at least one uppercase letter
    has_upper = any(ch.isupper() for ch in password)
    
    # Check for at least one lowercase letter
    has_lower = any(ch.islower() for ch in password)
    
    # Check for at least one digit
    has_digit = any(ch.isdigit() for ch in password)
    
    # Check for at least one special character from the allowed set
    has_special = any(ch in "@#$%^&*()-_=+?" for ch in password)
    
    # Check if the password contains any spaces
    has_space = " " in password 
    
    # Check if password length is more than 8
    long_enough = len(password) > 8

    # If password contains spaces, immediately print a warning and stop
    if has_space:
        print("Enter: Password should not contain spaces.")
        return

    # Print the validation checks for the user
    print(f"\nPassword entered: {password}")
    print("Length more than 8:" if long_enough else "not valid")
    print("Contains uppercase:" if has_upper else "not valid")
    print("Contains lowercase:" if has_lower else "not valid")
    print("Contains number:" if has_digit else "not valid") 
    print("Contains special character:" if has_special else "not valid") 
    
    # Final validation result: valid only if all conditions are met
    if all([has_upper, has_lower, has_digit, has_special, long_enough]) and not has_space:
        print("Password is valid!")
    else:
        print("Password is invalid.")

# Ask the user to enter a password
user_password = input("Enter your password: ")
validate_password(user_password)
```

# **QUESTION 23**
You are responsible for managing a study group for a course. Each student can be part of multiple study groups, and you need to track which groups a student is enrolled in. Students can join or leave study groups. What data structure would you choose to manage this efficiently, and why?

```python
#ANSWER 23
# Step 1: Create a dictionary to store students and their study groups
study_groups = {
    "John": ["Math", "Physics"],
    "Emily": ["Physics", "Biology"],
    "Alex": ["Math"]
}

# Step 2: Function to add (join) a group for a student
def join_group(student, group):
    # If student not in the dictionary, add them with an empty list
    if student not in study_groups:
        study_groups[student] = []
    # Add the group to the student's list if not already present
    if group not in study_groups[student]:
        study_groups[student].append(group)
        print(f"{student} has joined the {group} group.")
    else:
        print(f"{student} is already in the {group} group.")

# Step 3: Function to remove (leave) a group for a student
def leave_group(student, group):
    if student in study_groups and group in study_groups[student]:
        study_groups[student].remove(group)  # remove the group from the student's list
        print(f"{student} has left the {group} group.")
    else:
        print(f"{student} is not in the {group} group.")  # print if student or group doesn't exist

# Step 4: Function to display all groups a student is in
def show_groups(student):
    if student in study_groups:
        print(f"{student} is part of: {', '.join(study_groups[student])}")  # list all groups
    else:
        print(f"{student} is not enrolled in any groups yet.")

# --- Example usage ---
join_group("John", "Biology")    # John joins a new group
leave_group("Emily", "Physics")  # Emily leaves a group
show_groups("John")              # Show John's groups
show_groups("Emily")             # Show Emily's groups

# Display the entire data for reference
print("\nAll study group data:", study_groups)
```
